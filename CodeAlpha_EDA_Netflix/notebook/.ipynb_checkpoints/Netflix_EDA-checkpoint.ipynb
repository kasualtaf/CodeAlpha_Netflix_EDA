# Netflix Movies and TV Shows - Exploratory Data Analysis

## Introduction
In this project, we will perform Exploratory Data Analysis (EDA) on the Netflix Movies and TV Shows dataset. Our goal is to uncover insights about the content available on Netflix, including trends over time, top producing countries, and distribution of genres.

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for visualizations
sns.set_theme(style="darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## Load Dataset

```python
df = pd.read_csv('../dataset/netflix_titles.csv')
```

## Data Exploration

```python
# Display first 5 rows
display(df.head())

# Display last 5 rows
display(df.tail())
```

```python
# Check dataset shape
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Display column names
print("Column Names:")
print(df.columns.tolist())
```

```python
# Check data types
print("Data Types:")
print(df.dtypes)
```

```python
# Display summary statistics
display(df.describe(include='all'))
```

```python
# Check missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Check duplicate records
duplicates = df.duplicated().sum()
print(f"Number of duplicate records: {duplicates}")
```

## Data Cleaning

```python
# Remove duplicate rows (if any)
df.drop_duplicates(inplace=True)

# Handle missing values appropriately
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# Drop remaining rows with missing values in critical columns
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Convert "date_added" into datetime format
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')

# Extract "Year Added" from "date_added"
df['Year Added'] = df['date_added'].dt.year

display(df[['date_added', 'Year Added']].head())
```

## Exploratory Data Analysis

### 1. Number of Movies vs TV Shows (Pie Chart)

```python
type_counts = df['type'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'], explode=(0.05, 0))
plt.title('Proportion of Movies vs TV Shows', fontsize=15)
plt.show()
```

**Business Insight:** Movies make up nearly 70% of Netflix's content library, indicating a strong preference for producing or acquiring feature-length films over series.

### 2. Distribution of Ratings (Countplot)

```python
plt.figure(figsize=(10, 6))
order = df['rating'].value_counts().index
sns.countplot(data=df, x='rating', order=order, palette='crest')
plt.title('Distribution of Content Ratings', fontsize=15)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Business Insight:** The majority of content is rated TV-MA (Mature Audiences) and TV-14, showing that Netflix primarily targets adult and young adult demographics.

### 3. Top 10 Countries with Most Content (Bar Chart)

```python
df['primary_country'] = df['country'].apply(lambda x: x.split(',')[0])
top_countries = df[df['primary_country'] != 'Unknown']['primary_country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with Most Content', fontsize=15)
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

**Business Insight:** The United States is the largest producer of content on Netflix by a significant margin, followed by India and the United Kingdom.

### 4. Top 10 Genres (Bar Chart)

```python
df['primary_genre'] = df['listed_in'].apply(lambda x: x.split(',')[0])
top_genres = df['primary_genre'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='magma')
plt.title('Top 10 Genres on Netflix', fontsize=15)
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

**Business Insight:** Dramas, Comedies, and Documentaries are the most popular genres, indicating that Netflix focuses heavily on mainstream entertainment.

### 5. Content Added Over the Years

```python
content_by_year = df.groupby(['Year Added', 'type']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
content_by_year.plot(kind='line', marker='o', ax=plt.gca())
plt.title('Content Added Over the Years', fontsize=15)
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.legend(title='Type')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**Business Insight:** There was a rapid increase in content addition starting around 2015, peaking in 2019. The growth of Movies added was significantly steeper than TV Shows.

### 6. Distribution of Release Years (Histogram)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['release_year'], bins=30, kde=True, color='purple')
plt.title('Distribution of Release Years', fontsize=15)
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.xlim(1980, 2024)
plt.show()
```

**Business Insight:** Most of the content on Netflix was released in the last decade (2010 onwards), reflecting a preference for newer titles to keep the library fresh.

### 7. Release Year Distribution by Content Type (Boxplot)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='type', y='release_year', palette='pastel')
plt.title('Release Year Distribution by Content Type', fontsize=15)
plt.xlabel('Content Type')
plt.ylabel('Release Year')
plt.ylim(1980, 2024)
plt.show()
```

**Business Insight:** TV Shows tend to have more recent release years overall compared to Movies, with less variance in older content.

### 8. Top Directors

```python
top_directors = df[df['director'] != 'Unknown']['director'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='rocket')
plt.title('Top 10 Directors on Netflix', fontsize=15)
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

**Business Insight:** Rajiv Chilaka and Raúl Campos are among the top directors, highlighting a mix of international animation and stand-up comedy specials.

### 9. Correlation Heatmap (Heatmap)

```python
plt.figure(figsize=(8, 6))
corr_matrix = df[['release_year', 'Year Added']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap', fontsize=15)
plt.show()
```

**Business Insight:** There is a positive correlation between 'release_year' and 'Year Added', meaning newer releases tend to be added to Netflix shortly after their release.

## Key Insights
1. **Content Type**: Movies make up nearly 70% of Netflix's content library.
2. **Top Producing Country**: The United States is the leading content producer, followed by India.
3. **Popular Genres**: Dramas, Comedies, and Documentaries are the most common genres.
4. **Target Audience**: Most content is tailored for mature audiences (TV-MA).
5. **Content Age**: Netflix focuses on recent content, with the majority of titles released post-2010.

## Conclusion
This Exploratory Data Analysis provides a comprehensive overview of Netflix's content library. We observed that Netflix is heavily invested in modern, mature-rated movies and expanding its global reach, particularly in the US and India. The platform focuses heavily on acquiring newer titles to keep audiences engaged.
